```python
import tensorflow as tf
import numpy as np
```

```python
x_train = []
y_train =[]

for x in range(1, 100):
    x_train.append(x)
    y_train.append(x*2+1)

print(x_train)
print(y_train)
```

```python
from tensorflow.keras.optimizers import SGD, Adam

# y = wx + b
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1, input_dim = 1))

# 학습율(learning_rate) : 학습이 진행되는 동안의 보폭
# decay : learning_rate_decay
adam = Adam(learning_rate=0.01, decay=1e-5, beta_1=0.99, beta_2=0.99)

model.compile(loss='mse',
              optimizer=adam)

history_adam = model.fit(x_train,y_train,epochs=500, batch_size=8)
```

```python
# Adam의 기본값으로 설정한 후 예측값
# y =2x+1 => 7
# loss: 6.3184e-05
print(model.predict(np.array([3])))
```

```python
from tensorflow.keras.optimizers import SGD

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1, input_dim = 1))

sgd = SGD(learning_rate=0.0001, decay=1e-7, momentum=0.9)

model.compile(loss='mse',
              optimizer=sgd)

history_sgd = model.fit(x_train,y_train,epochs=500, batch_size=8)
```

**학습율(Learning Rate)** 및 **학습에 따른 보폭 감소(Learning Rate Decay)**에 대한 **Loss값 변화**

```python
import matplotlib.pyplot as plt

def plot_model(history):

    # summarize history for loss
    plt.rcParams["figure.figsize"] = (10,10)
    plt.plot(history_adam.history['loss'], 'c')
    plt.plot(history_sgd.history['loss'], '--')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['adam','sgd'], loc='upper right')
    plt.show()

plot_model(model.history)
```

```python
print(model.predict(np.array([3])))
```

```python
# callback함수를 통해서 learning_rate를 줄일 경우
# 'loss'를 기준으로 10epoch동안 loss값이 줄어들지 않을 경우 기존의 learning_rate를 반으로 줄인다.
callback = tf.keras.callbacks.ReduceLROnPlateau(monitor='loss', factor=0.5,patience=10)

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(1, input_dim = 1))


model.compile(loss='mse', optimizer=SGD(learning_rate=0.0001, momentum=0.9))

history_cb = model.fit(x_train,y_train, epochs=500, batch_size=8, callbacks=[callback])
```

```python
print(model.predict(np.array([3])))
```
